```python
for group1 in df[df.Body_Type.isnull()].mfg.unique().tolist():
    for group2 in df[df.Body_Type.isnull()]['Nr.OfDoors'].unique().tolist():
        for group3 in df[df.Body_Type.isnull()]['Nr.OfSeats'].unique().tolist():
        
            cond1 = (df['mfg']==group1)
            cond2 = (df['mfg']==group1) & (df['Nr.OfDoors']==group2)
            cond3 = (df['mfg']==group1) & (df['Nr.OfDoors']==group2) & (df['Nr.OfSeats']==group3)
        
            mode1 = list(df[cond1]['Body_Type'].mode())
            mode2 = list(df[cond2]['Body_Type'].mode())
            mode3 = list(df[cond3]['Body_Type'].mode())
        
            if mode3 != []:
                df.loc[cond3, 'Body_Type'] = df.loc[cond3, 'Body_Type'].fillna(df[cond3]['Body_Type'].mode()[0])
            elif mode2 != []:
                df.loc[cond3, 'Body_Type'] = df.loc[cond3, 'Body_Type'].fillna(df[cond2]['Body_Type'].mode()[0])
            elif mode1 != []:
                df.loc[cond3, 'Body_Type'] = df.loc[cond3, 'Body_Type'].fillna(df[cond1]['Body_Type'].mode()[0])    
            else:
                df.loc[cond3, 'Body_Type'] = df.loc[cond3, 'Body_Type'].fillna(df['Body_Type'].mode()[0])

'''this part is for rows that still cannot be filled due to nan values in other columns'''
for group1 in df[df.Body_Type.isnull()].mfg.unique().tolist():
    
    cond1 = (df['mfg']==group1)
    
    df.loc[cond1, 'Body_Type'] = df.loc[cond1, 'Body_Type'].fillna(df[cond1]['Body_Type'].mode()[0])   
```
